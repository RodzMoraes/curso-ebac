## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

# Método 1
txt_count = cancao.count('roda')
print(txt_count)

# Método 2
txt_index = cancao.index('gigante')
print(txt_index)

# Método 3
txt_rpartition = cancao.rpartition('gigante')
print(txt_rpartition)
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

selic_inicio = noticia.index('2')
selic_final = noticia.index('%')
# Aqui eu pensei: melhor do que ficar contando, posso usar a lógica da programação
# a meu favor, e armazenar em uma variável o índice correspondente ao começo
# do texto que eu quero, e armazenar em outra variável o índice correspondente
# ao final desse texto, e usar essas variáveis para definir o início e o final
# do slice.

selic = noticia[selic_inicio:(selic_final+1)]
print(selic)

ano_index = noticia.index('6')
ano = noticia[ano_index]
print(ano)
```
