```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests

%matplotlib notebook

sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
idade_mae = sinasc['IDADEMAE']

plt.figure()
plt.boxplot(idade_mae)
plt.xlabel("Idade da Mãe")
plt.ylabel("Idade")
plt.title("Boxplot da Idade da Mãe")
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
peso_bebe = sinasc['PESO']

plt.figure()
plt.boxplot(peso_bebe)
plt.xlabel("Peso do Bebê")
plt.ylabel("Peso (g)")
plt.title("Boxplot do Peso do Bebê")
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
idade_mae = sinasc['IDADEMAE']

plt.figure()
plt.hist(idade_mae, bins=10, edgecolor='black')
plt.xlabel("Idade da Mãe")
plt.ylabel("Frequência")
plt.title("Histograma da Idade da Mãe")
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
peso_bebe = sinasc['PESO']

plt.figure()
plt.hist(peso_bebe, bins=10, edgecolor='black')
plt.xlabel("Peso do Bebê")
plt.ylabel("Frequência")
plt.title("Histograma do Peso do Bebê")
plt.show()
```
