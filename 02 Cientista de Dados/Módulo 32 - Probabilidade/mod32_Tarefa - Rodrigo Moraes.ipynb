# Probabilidade - Tarefa

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
from scipy.stats import norm, t
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
media = 15 
desvio_padrao = 0.1
limite_superior = 15.2

probabilidade = 1 - norm.cdf(limite_superior, loc=media, scale=desvio_padrao)
print(f"A probabilidade de um pacote ter mais de {limite_superior} kg  de {probabilidade:.2%}")
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
media = 15
desvio_padrao = 0.1
percentil = 0.95

limite_inferior = norm.ppf(percentil, loc=media, scale=desvio_padrao)
print(f"O percentil de ordem {percentil*100}% para o peso dos pacotes  de {limite_inferior:.2f} kg")
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y = np.linspace(-6, 6, 200)
# Seu cdigo comea aqui

# Distribuio Normal Padro (Y1)
prob_Y1 = norm.cdf(Y, loc=0, scale=1)

# Distribuio t-student com 5 graus de liberdade (Y2)
prob_Y2 = t.cdf(Y, df=5)

for i in range(len(Y)):
    print(f"Para Y = {Y[i]}:")
    print(f"Probabilidade para Y1: {prob_Y1[i]:.4f}")
    print(f"Probabilidade para Y2: {prob_Y2[i]:.4f}")
    print()
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
plt.plot(Y, probabilidade_Y1, label='Y1 - Normal Padro')
plt.plot(Y, probabilidade_Y2, label='Y2 - t-Student (5 graus de liberdade)')

plt.xlabel('Y')
plt.ylabel('Probabilidade')
plt.title('Probabilidade de ser menor que Y para Y1 e Y2')
plt.legend()

plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
P = np.linspace(0, 1, 20)
# Seu cdigo comea aqui

# Distribuio Normal Padro (Y1)
percentis_Y1 = norm.ppf(P, loc=0, scale=1)

# Distribuio t-student com 5 graus de liberdade (Y2)
percentis_Y2 = t.ppf(P, df=5)

for i in range(len(P)):
    print(f"Para P = {P[i]:.2f}:")
    print(f"Percentil para Y1: {percentis_Y1[i]:.4f}")
    print(f"Percentil para Y2: {percentis_Y2[i]:.4f}")
    print()
```

6. Faa um grfico com os dados do item anterior.

```python
plt.plot(P, percentis_Y1, label='Y1 - Normal Padro')
plt.plot(P, percentis_Y2, label='Y2 - t-Student (5 graus de liberdade)')

plt.xlabel('P')
plt.ylabel('Percentil')
plt.title('Percentis para Y1 e Y2')
plt.legend()

plt.show()
```
