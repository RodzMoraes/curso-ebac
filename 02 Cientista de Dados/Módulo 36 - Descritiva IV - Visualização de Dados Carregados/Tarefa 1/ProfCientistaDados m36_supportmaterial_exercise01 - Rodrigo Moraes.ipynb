````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_pickle('base_nfp.pkl')
df.head()
````

## Anlise da probabilidade de retorno

Considere somente os dados de janeiro de 2020 em diante para esta anlise.

- Analise a proporo de notas que tiveram retorno>0 por categoria de estabelecimento. Esta anlise pode ser por tabela ou grfico.

````python
df.rename(columns={'Data Emisso':'Data_Emisso'}, inplace=True)
df.head(2)
````

````python
df.set_index('Data_Emisso', inplace=True)
df_cut = df['2020-01-01':]

df_cut.dropna(inplace=True)
df_cut.reset_index(inplace=True)
df_cut.head(2)
````

````python
fig, ax = plt.subplots(figsize=(13,7))
sns.scatterplot(ax=ax, data=df_cut[(df_cut['Retorno'] > 0) & (df_cut['Retorno'] < 200) ], x='categoria', y='Retorno')
````

Pelo grfico podemos notar claramente que ``Mercado`` possui valores altos de **Retorno** e ``Auto posto`` possui valores bem baixos.

````python
fig, ax = plt.subplots(figsize=(13,7))
sns.countplot(ax=ax, data=df_cut, x='categoria', hue='flag_credito')
````

Neste grfico observamos que ``Alimentos`` e ``Restaurante`` possuem uma proporo maior de valores 1 na varivel **flag_credito**, enquanto ocorre o oposto com os demais.

## Clculo do WOE

- Calcule o WOE das categorias, sendo que o evento em questo  a nota ter retorno>0.
- Analise este WOE por tabela e por grfico.

````python
tab = pd.crosstab(df_cut.categoria, df_cut.flag_credito, margins=True, margins_name='total')
tab['bad_rate'] = tab[1]/tab['total']
tab['odds'] = tab[1]/tab[0]
tab['odds_ratio'] = tab['odds']/tab.loc['total', 'odds']
tab['woe'] = np.log(tab.odds_ratio)
tab
````

````python
fig, ax = plt.subplots(figsize=(13,7))
sns.lineplot(ax=ax, data=tab, x='categoria', y='woe', marker='o')
````

## Clculo do *Information Value*

Calcule o *IV* dessa varivel.

````python
tab = pd.crosstab(df_cut.categoria, df_cut.flag_credito, margins=True, margins_name='total')
tab['pct_evento'] = tab[tab.columns[0]]/tab.loc['total',tab.columns[0]]
tab['pct_nao_evento'] = tab[tab.columns[1]]/tab.loc['total',tab.columns[1]]
tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento)*tab.woe
print(tab.iv_parcial.sum())
tab
````
