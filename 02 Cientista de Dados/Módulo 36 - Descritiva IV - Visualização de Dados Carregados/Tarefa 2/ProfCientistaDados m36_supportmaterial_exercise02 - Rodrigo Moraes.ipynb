## Criando variveis

**Variveis temporais**:  possvel que haja alguma regra de liberao de crditos de acordo com o dia do ms, o dia da semana, fim de semana (sim ou no), ou at alteraes temporais da regra. Vamos criar as seguintes variveis para testar estas hipteses:

- Dia da semana
- Dummy para fim de semana
- Dia do ms
- Ms (j na base)
- Trimestre
- Ano (j na base)

**Valor da nota**: Vamos criar tambm algumas variveis para resumir o valor da nota, e posteriormente criar uma intuio sobre categorizaes de variveis contnuas.

- Crie categorizaes da varivel ```Valor NF```, por quants, em 5, 10, 20 e 50 categorias.

````python
# Dia da semana
df_cut['Dia_semana'] = df_cut['Data_Emisso'].dt.day_name()

# Dummy para fim de semana
df_cut['Fim_de_semana'] = ((df_cut['Data_Emisso'].dt.dayofweek) // 5 == 1).astype(float)

# Dia do ms
df_cut['Dia_mes'] = df_cut['Data_Emisso'].dt.day

# Ms
df_cut['Mes'] = df_cut['Data_Emisso'].dt.month

# Trimestre
df_cut['Trimestre'] = df_cut['Data_Emisso'].dt.quarter

# Ano
df_cut['Ano'] = df_cut['Data_Emisso'].dt.year
````

````python
df_cut['Valor_NF_5_cat'] = pd.qcut(x=df_cut['Valor NF'], q=5)
df_cut['Valor_NF_10_cat'] = pd.qcut(x=df_cut['Valor NF'], q=10)
df_cut['Valor_NF_20_cat'] = pd.qcut(x=df_cut['Valor NF'], q=20)
df_cut['Valor_NF_50_cat'] = pd.qcut(x=df_cut['Valor NF'], q=50)
````

## Calcule o IV de todas essas variveis

Sugesto: Construa um metadados e tente utilizar uma funo como a que foi utilizada durante a aula. Voc vai conseguir fazer esse clculo para as variveis uma a uma, mas em casos reais em que vamos mais alm na criao de variveis, podemos ter centenas, ou at milhares de variveis, esses clculos *precisam* ser automticos nesses casos.

````python
df_cut.info()
````

````python
# Definindo metadados
df_short = df_cut.drop(columns=['Data_Emisso', 'CNPJ emit.', 'Emitente', 'No.', 'Valor NF', 'Data Registro', 'Crditos', 
                     'Situao do Crdito', 'Retorno'])

metadados = pd.DataFrame({'dtype':df_short.dtypes})
metadados
````

````python
# Atribuindo papis
metadados['papel'] = 'covariavel'
metadados.loc['flag_credito','papel'] = 'resposta'
metadados['nunique'] = df_cut.nunique()
metadados
````

````python
# Definindo a funo

def IV(variavel, resposta):
    tab = pd.crosstab(variavel, resposta, margins=True, margins_name='total')

    rtulo_evento = tab.columns[0]
    rtulo_nao_evento = tab.columns[1]

    tab['pct_evento'] = tab[rtulo_evento]/tab.loc['total',rtulo_evento]
    tab['pct_nao_evento'] = tab[rtulo_nao_evento]/tab.loc['total',rtulo_nao_evento]
    tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
    tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento)*tab.woe
    return tab['iv_parcial'].sum()
````

````python
# Populando metadados

for var in metadados[metadados.papel=='covariavel'].index:
    if  (metadados.loc[var, 'nunique']>50):
        metadados.loc[var, 'IV'] = IV(pd.qcut(df_short[var],5,duplicates='drop'), df_short.flag_credito)
    else: 
        metadados.loc[var, 'IV'] = IV(df_short[var], df_short.flag_credito)
    
metadados
````

## Avalie o WOE e o IV no tempo

Avalie o WOE e o IV no tempo da varivel ```categoria```.

````python
df_cut_3['pct_evento']     = df_cut_3[1]/df_cut_3[1].groupby(level=0).sum()
df_cut_3['pct_nao_evento'] = df_cut_3[0]/df_cut_3[0].groupby(level=0).sum()
df_cut_3['WOE'] = np.log(df_cut_3.pct_evento/df_cut_3.pct_nao_evento)
df_cut_3.head()
````

````python
fig, ax = plt.subplots(figsize=(13,7))

sns.pointplot(data=df_cut_3.reset_index(), y = 'WOE', x = 'Trimestre', hue='categoria', ax=ax);

ymax = max(2.1, max(df_cut_3.WOE))
ymin = min(-2.1, min(df_cut_3.WOE))
ax.set_ylim([ymin, ymax])

# Linhas de referncia
ax.axhline(2, linestyle=':'  , color='red', alpha=.5)
ax.axhline(1, linestyle=':'  , color='red', alpha=.5)
ax.axhline(0, linestyle='--' , color='grey', alpha=.5)
ax.axhline(-1, linestyle=':' , color='red', alpha=.5)
ax.axhline(-2, linestyle=':' , color='red', alpha=.5)

ax.set_title(f'WOE ao longo dos trimestres para varivel categorias')
ax.legend(bbox_to_anchor=(.85, -.1), ncol=5)
````

````python
df_results = df_cut_3.reset_index()
df_results['WOE'] = df_results['WOE'].round(6)

table_results = df_results.pivot(index='categoria', columns='Trimestre', values='WOE')
table_results.columns = ['Trimestre ' + str(col) for col in table_results.columns]
table_results = table_results.reset_index()

table_results = table_results.rename_axis(None, axis=1)
table_results = table_results.rename(columns={'categoria': 'Categoria'})

print(table_results.to_string(index=False))
````

````python
def plot_iv(IV):
    fig, ax = plt.subplots()
    x = IV.index
    a = [.02]*len(IV.index)
    b = [.1]*len(IV.index)
    c = [.3]*len(IV.index)
    d = [.5]*len(IV.index)
    e = [.6]*len(IV.index)

    ax.fill_between(IV.index, a, color='grey', alpha=.2)
    ax.fill_between(IV.index, a, b, color='orange', alpha=.1)
    ax.fill_between(IV.index, b, c, color='green', alpha=.1)
    ax.fill_between(IV.index, c, d, color='blue', alpha=.1)
    ax.fill_between(IV.index, d, e, color='purple', alpha=.1)
    ax.plot(IV)

    ax.set_title("Information Value de categoria ao longo dos Trimestres")
    ax.set_ylabel("Information Value")
    ax.set_xlabel("Trimestres")
````

````python
plot_iv(((df_cut_3.pct_evento - df_cut_3.pct_nao_evento)*df_cut_3.WOE).groupby(level=0).sum())
````
